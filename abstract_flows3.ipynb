```python
#libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import folium
import os
import sys
sys.path.append(os.path.abspath('abstract_flows'))
import arrow
import grid
import flows
```

```python
dir = '/home/user/Desktop/Datos/'
#dir = '/Users/antoniomendez/Desktop/Tesis/Datos/datos_limpios/'
```

```python
estaciones_eco = np.load(dir + 'Adj_eco/matrices_estaciones/est_2019.npy')
```

```python
data_eco = pd.read_csv(dir + 'ecobici/ecobici_2019.csv')
data_eco = data_eco[data_eco['Fecha_Retiro'].str.contains('2019-01')]
data_eco.head()  
```

```python
data_eco.tail()
```

```python
# puede tardar hasta media hora para 15 dias
for i in range(1,32):
    print('Dia ' + str(i))
    current_data = data_eco[data_eco['Fecha_Retiro'].str.contains('2019-01-' + str(i).zfill(2))]
    current_counter = flows.count_trips_ecobici(current_data, threshold=1)
    current_stations = np.unique(np.concatenate(current_data[['Ciclo_Estacion_Retiro', 'Ciclo_Estacion_Arribo']].values))
    current_grid = grid.Grid(18, 10, 'ecobici')
    current_map = current_grid.map_around()
    current_station_cells = flows.stations_and_cells(current_grid.geodataframe(), current_stations, estaciones_eco)
    current_graph = flows.abstract_flows(current_counter, current_grid.geodataframe(), current_station_cells, estaciones_eco)
    current_map = flows.plot_abstracted_graph(current_graph, current_map, title=f'Flows Ecobici: 2019-01-{i:02d}')
    current_map.save(dir + 'viajes_eco/2019-01-' + str(i).zfill(2) + '.html')
```

```python
estaciones_mibici = np.load(dir + 'Adj_mibici/matrices_estaciones/est_2019.npy')
```

```python
data_mibici = pd.read_csv(dir + 'mibici/mibici_2019.csv')
data_mibici = data_mibici[data_mibici['Inicio_del_viaje'].str.contains('2019-01')]
data_mibici.head()
```

```python
data_mibici.tail()
```

```python
# puede tardar hasta 10 minutos para 15 dias
for i in range(1,32):
    print('Dia ' + str(i))
    current_data = data_mibici[data_mibici['Inicio_del_viaje'].str.contains('2019-01-' + str(i).zfill(2))]
    current_counter = flows.count_trips_mibici(current_data, threshold=1)
    current_stations = np.unique(np.concatenate(current_data[['Origen_Id', 'Destino_Id']].values))
    current_grid = grid.Grid(10, 10, 'mibici')
    current_map = current_grid.map_around()
    try:
        current_station_cells = flows.stations_and_cells(current_grid.geodataframe(), current_stations, estaciones_mibici)
        current_graph = flows.abstract_flows(current_counter, current_grid.geodataframe(), current_station_cells, estaciones_mibici)
        current_map = flows.plot_abstracted_graph(current_graph, current_map, title=f'Flows Mibici: 2019-01-{i:02d}')
    except Exception as e:
        print(e)
    current_map.save(dir + 'viajes_mibici/2019-01-' + str(i).zfill(2) + '.html')
```
