```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
import os
from datetime import datetime, timedelta
from matplotlib.colors import LogNorm
```

```python
def leer_matriz(nombre_archivo):
    matriz = []
    with open(nombre_archivo, 'r') as archivo:
        archivo.readline()
        archivo.readline()
        for linea in archivo:
            fila = [float(valor) for valor in linea.strip().split()]
            matriz.append(fila)
    return np.array(matriz)
```

```python
# Lista de años
years = [2019, 2020, 2021, 2022, 2023, 2024]

# Lista para almacenar las matrices y las fechas correspondientes
matrices = []
dates = []

# Iterar sobre cada año
for year in years:
    # Definir la fecha de inicio y fin para cada año
    start_date = datetime(year, 1, 1)
    end_date = datetime(year, 12, 31)

    # Incremento de un día
    delta = timedelta(days=1)
    current_date = start_date

    # Iterar sobre cada día del año
    while current_date <= end_date:
        date_str = current_date.strftime('%d_%m_%Y')
        filename = f'Adj_eco/matrices_{year}/{date_str}.txt'

        # Verificar si el archivo existe
        if os.path.exists(filename):
            # Leer la matriz desde el archivo
            matrix = leer_matriz(filename)
            matrices.append(matrix)
            dates.append(current_date)
        else:
            print(f"El archivo {filename} no existe.")
        current_date += delta

# Verificar si se cargaron matrices
if not matrices:
    raise ValueError("No se encontraron matrices para generar la animación.")

# Obtener el valor máximo para escalar los colores del heatmap
vmax = np.max([np.max(matrix) for matrix in matrices])

# Crear la figura y el primer heatmap
fig, ax = plt.subplots()
heatmap = ax.imshow(matrices[0], cmap='viridis', vmin=0, vmax=vmax)

# Función para actualizar el heatmap en cada frame
def update(frame):
    ax.clear()
    ax.set_title(f"Fecha: {dates[frame].strftime('%d/%m/%Y')}")
    heatmap = ax.imshow(matrices[frame], cmap='hot', norm=LogNorm(vmin=1e-3, vmax=vmax))
    return [heatmap]

# Crear la animación
ani = FuncAnimation(fig, update, frames=len(matrices), interval=150)

# Guardar la animación o mostrarla
ani.save("evolucion_conexiones_ecobici.mp4", writer="ffmpeg", codec='mpeg4')
# plt.show()
```
