```python
import numpy as np
import pandas as pd
```

```python
def leer_matriz(nombre_archivo):
    matriz = []
    with open(nombre_archivo, 'r') as archivo:
        archivo.readline()
        archivo.readline()
        for linea in archivo:
            fila = [float(valor) for valor in linea.strip().split()]
            matriz.append(fila)
    return matriz

def escribir_matriz(matriz, archivo):
    with open(archivo, 'w') as file:
        file.write(f"{len(matriz)} {len(matriz[0])}\n\n")
        for fila in matriz:
            fila_str = ' '.join(str(elemento) for elemento in fila)
            file.write(fila_str + '\n')
```

```python
'''Para crear las matrices de estaciones que incluyan su ubicación espacial

Esto siguiente es para mibici'''
estaciones_mibici = pd.read_csv('nomenclaturas/nomenclatura_2024_07.csv', encoding='latin1')
estaciones_mibici.head()
```

```python
years = ['2019', '2020', '2021', '2022', '2023', '2024']
for year in years:
    x = np.array(leer_matriz(f'Adj_mibici/matrices_estaciones/estaciones_{year}.txt'))
    est = x[:, 0]
    estaciones = []
    for i in range(len(est)):
        lat = estaciones_mibici[estaciones_mibici['id'] == est[i]]['latitude'].values[0]
        lon = estaciones_mibici[estaciones_mibici['id'] == est[i]]['longitude'].values[0]
        estaciones.append([int(est[i]), lat, lon])
    estaciones = np.array(estaciones)
    escribir_matriz(estaciones, f'Adj_mibici/matrices_estaciones/est_{year}.txt')
```
