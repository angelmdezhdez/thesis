```python
n_fixed = pd.read_csv('nomenclaturas/nomenclatura_2024_07.csv', encoding='latin1')
n_fixed.head()
```

```python
ns = []
ns.append(pd.read_csv('nomenclaturas/nomenclatura_2022_11.csv', encoding='latin1'))
ns.append(pd.read_csv('nomenclaturas/nomenclatura_2023_02.csv', encoding='latin1'))
ns.append(pd.read_csv('nomenclaturas/nomenclatura_2023_10.csv', encoding='latin1'))
```

```python
for i in range(len(ns)):
    print('Revisando nomenclatura', i)
    for id in ns[i]['id']:
        current_data_fixed = n_fixed[n_fixed['id'] == id]
        current_data = ns[i][ns[i]['id'] == id]
        if current_data_fixed['latitude'].values != current_data['latitude'].values:
            print(id)
            print('latitud en 2024: ', current_data_fixed['latitude'].values, 'latitud en ', i, ': ', current_data['latitude'].values)
        if current_data_fixed['longitude'].values != current_data['longitude'].values:
            print(id)
            print('longitud en 2024: ', current_data_fixed['longitude'].values, 'longitud en ', i, ': ', current_data['longitude'].values)
```

```python
for i in range(len(ns)):
    print('Revisando nomenclatura', i)
    for id in ns[i]['id']:
        current_data_fixed = n_fixed[n_fixed['id'] == id]
        current_data = ns[i][ns[i]['id'] == id]
        if current_data_fixed['latitude'].values != current_data['latitude'].values or current_data_fixed['longitude'].values != current_data['longitude'].values:
            print('ID: ', id)
            print('latitud en 2024: ', current_data_fixed['latitude'].values, 'latitud en ', i, ': ', current_data['latitude'].values)
            print('longitud en 2024: ', current_data_fixed['longitude'].values, 'longitud en ', i, ': ', current_data['longitude'].values)
            # para verificar que en los datos de 2024 haya alguno con la misma latitud
            matching_latitude = n_fixed[n_fixed['latitude'] == current_data['latitude'].values[0]]

            if not matching_latitude.empty:
                for match_id in matching_latitude['id'].values:
                    match_data = n_fixed[n_fixed['id'] == match_id]
                    if not match_data.empty and match_data['longitude'].values[0] == current_data['longitude'].values[0]:
                        print("ID en n_fixed con la misma latitud y longitud:", match_id)
            else:
                print("No hay coincidencias en latitud")
       
            # Verificar coincidencias en latitud
            matching_longitude = n_fixed[n_fixed['longitude'] == current_data['longitude'].values[0]]
            if not matching_longitude.empty:
                for match_id in matching_longitude['id'].values:
                    match_data = n_fixed[n_fixed['id'] == match_id]
                    if not match_data.empty and match_data['latitude'].values[0] == current_data['latitude'].values[0]:
                        print("ID en n_fixed con la misma latitud y longitud:", match_id)
            else:
                print("No hay coincidencias en longitud")
```
